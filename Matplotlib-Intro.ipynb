## Matplotlb Example Workflow

```python
# 0. Import Matplotlib Library and modules.

%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare Data 

x = [1,2,3,4,5]
y = [11,22,33,44,55]

# Set Up Plot

fig , ax = plt.subplots(figsize=(10,10))

# Plot Data
ax.plot(x,y)

# Customize PLot
ax.set(title="Simple Plot", 
       xlabel="x-axis", 
       ylabel="y-axis")

# Save and Show

fig.savefig('./numpy-images/sample-plot.png')

```

### Heart Disease Section Plotting pyplot and OO method Matplotlib

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# pyplot method
over_50.plot(kind='scatter',x='age',y='chol',c='target',figsize=(10,10));
```

```python
# OO Method (Object Oriented)
fig , ax =plt.subplots(figsize=(10,6))
# pyplot method
over_50.plot(kind='scatter',x='age',y='chol',c='target',cmap='terrain',ax=ax);ax.set_xlim([45,100])
```

```python
## OO Method from Scratch 
fig , ax =plt.subplots(figsize=(16,6))
# Plot the Data
scatter = ax.scatter(x = over_50["age"],
                 y=over_50["chol"],
                 c=over_50["target"],cmap='winter');
# Customize the plot
ax.set(title="Heart Disease And Cholestrol levels",xlabel="Age",ylabel="Cholestrol") 
ax.legend(*scatter.legend_elements(),title="Target")
ax.axhline(over_50["chol"].mean(),linestyle='--')
```

```python
over_50.head()
```

```python
fig ,(ax0,ax1) = plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)
plt.style.use('seaborn')
scatter = ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"],cmap='winter')
ax0.set(title="Heart Disease And Cholestrol levels",ylabel="Cholestrol") 
ax0.legend(*scatter.legend_elements(),title="Target")
ax0.axhline(over_50["chol"].mean(),linestyle='--')

scatter = ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"],cmap='winter')
ax1.set(title="Heart Disease And Thalach levels",xlabel="Age",ylabel="Thalach") 
ax1.legend(*scatter.legend_elements(),title="Target")
ax1.axhline(over_50["thalach"].mean(),linestyle='--')

fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight='bold');
```

```python
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize

# Assuming `pokemon_csv` is your DataFrame
c = pokemon_csv["Generation"]
x = pokemon_csv["Type 1"]
y = pokemon_csv["Attack"]

# Normalize the color values
norm = Normalize(vmin=min(c), vmax=max(c))

# Create scatter plot using Matplotlib
plt.figure(figsize=(16, 6))
scatter = plt.scatter(x, y, c=c, cmap='winter', norm=norm)

# Add a colorbar
plt.colorbar(scatter, label="Generation")

# Add labels and title
plt.xlabel("Type 1")
plt.ylabel("Attack")
# plt.xlim(0, 250)  # For numeric encoding of 'Type 1'
plt.ylim(0, 200)  # Example y-axis limit, adjust based on your data

plt.title("Scatter Plot of Type 1 vs Attack with Generation Coloring")
plt.show()
```
