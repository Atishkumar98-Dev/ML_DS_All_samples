```python
# 1. Get the Data Ready
import pandas as pd
import numpy as np
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
# Create X (Features Matrix) 

# NOTE : Dont Forget to mention Axis When Dropping
X = heart_disease.drop("target",axis=1)

# Create Labels
y = heart_disease["target"]
```

```python
# Choose the estimator/algorithm for our problems 
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=100)
clf.get_params()
```

```python
# Fit the model to the Training Data.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
clf.fit(X_train,y_train)
```

```python
# make a prediction
y_label = clf.predict(np.array([0,1,2,3,4]))
```

```python
y_preds = clf.predict(X_test)
y_preds
```

```python
clf.score(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
from sklearn.metrics import classification_report , confusion_matrix, accuracy_score

print(classification_report(y_test,y_preds))
```

```python
print(confusion_matrix(y_test,y_preds))
```

```python
print(accuracy_score(y_test,y_preds))
```

```python
y_test
```

```python
y_preds
```

```python
# Improve a Model 
np.random.seed(42)
for i in range(10,100,10):
    print(F"Trying model with {i} estimators...")
    clf = RandomForestClassifier(n_estimators=i).fit(X_train,y_train)
    print(F"Model accuracy on test set:{clf.score(X_test,y_test)*100:2f}%")
    print("")
```

```python
import pickle
pickle.dump(clf,open("random_forest_model_1.pkl","wb"))
```

```python
loaded_model = pickle.load(open("random_forest_model_1.pkl","rb"))
loaded_model.score(X_test,y_test)
```

```python
car_sales = pd.read_csv("car-sales-extended.csv")
car_sales.head()
```

```python
len(car_sales)
```

```python
X = car_sales.drop("Price",axis = 1)
y = car_sales["Price"]
```

```python
X_train, X_test, y_train , y_test = train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
y_test.shape
```

```python
# Model Training 
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X_train, y_train)
model.score(X_test,y_test)
```

```python
# Convert strings/categories to numbers
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_feature = ["Make", "Colour","Doors"]
one_hot = OneHotEncoder()
transformer  = ColumnTransformer([('one_hot',one_hot, categorical_feature,)],remainder="passthrough")
transformed_X = transformer.fit_transform(X)
transformed_X
```

```python
# Via Pandas get dummy numbers
```

```python
dummies = pd.get_dummies(car_sales[["Make","Colour", "Doors"]])
dummies
```

```python
# Refitting Model
```

```python
np.random.seed(42)
X_train, X_test, y_train, y_test = train_test_split(transformed_X, y , test_size = 0.2)
model.fit(X_train,y_train)
model.score(X_test,y_test)
```
